```python
%%time
reviews= {'hotel':[],'commentaire' : [], 'note': []}
page = 0
while True:
    try:
        page += 1
        container = browser.find_elements_by_class_name('location-review-review-list-parts-RatingLine__bubbles--GcJvM')
        
        
        for j in range(len(container)):
            container = browser.find_elements_by_class_name('location-review-review-list-parts-SingleReview__mainCol--1hApa')

            scrap_review= container[j].find_element_by_css_selector('.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC span')
            


            container2=browser.find_elements_by_class_name('location-review-review-list-parts-RatingLine__bubbles--GcJvM')

            rating_a = container2[j].find_element_by_xpath(".//span[contains(@class, 'ui_bubble_rating bubble_')]")
            rating_b1=rating_a.get_attribute("class")
#                 print(rating_b1)
            rating_b2 = rating_b1.split("_")
            rating = rating_b2[3]
            if rating is not None:
                reviews['note'].append(rating)
                reviews['commentaire'].append(scrap_review.text)
                reviews['hotel'].append('Novotel Paris Les Halles')
            else:
                print('pass')
            time.sleep(1)
            
        print(len(reviews['commentaire'])== len(reviews['note']),len(reviews['commentaire']),len(reviews['note']),page)
        arrow = browser.find_element_by_xpath("//*[text() = 'Suivant']").click()
        time.sleep(3)
    except StaleElementReferenceException as e:
            print(page)
            print(e)
#     except NoSuchElementException :
#         browser.close()
    except ElementClickInterceptedException:
#         browser.close()
        break
```

```python
df=pd.DataFrame(reviews,columns=['hotel','commentaire','note'])
# df.shape
df.to_csv("Novotel_Paris_reviews.csv")
```

```python
def get_data_hotel(url):    
    
    browser = webdriver.Chrome()
    browser.get(url)
    browser.maximize_window()
    
    name_hotel=re.findall(r'Reviews-([A-z-0-9_]+)\-',url) 
    reviews= {'hotel':[],'commentaire' : [], 'note': []}
    page = 0
    while True:
        try:
            page += 1
            container = browser.find_elements_by_class_name('location-review-review-list-parts-RatingLine__bubbles--GcJvM')


            for j in range(len(container)):
                container = browser.find_elements_by_class_name('location-review-review-list-parts-SingleReview__mainCol--1hApa')

                scrap_review= container[j].find_element_by_css_selector('.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC span')

                container2=browser.find_elements_by_class_name('location-review-review-list-parts-RatingLine__bubbles--GcJvM')

                rating_a = container2[j].find_element_by_xpath(".//span[contains(@class, 'ui_bubble_rating bubble_')]")
                rating_b1=rating_a.get_attribute("class")
    #                 print(rating_b1)
                rating_b2 = rating_b1.split("_")
                rating = rating_b2[3]
                if rating is not None:
                    reviews['note'].append(rating)
                    reviews['commentaire'].append(scrap_review.text)
                    reviews['hotel'].append(name_hotel[0])
                else:
                    print('pass')
                time.sleep(1)

            print(len(reviews['commentaire'])== len(reviews['note']),len(reviews['commentaire']),len(reviews['note']),page)
            arrow = browser.find_element_by_xpath("//*[text() = 'Suivant']").click()
            time.sleep(3)
            
        except StaleElementReferenceException as e:
                print(page)
                print(e)
    #     except NoSuchElementException :
    #         browser.close()
        except ElementClickInterceptedException:
            browser.close()
            break

         
    df=pd.DataFrame(reviews,columns=['hotel','commentaire','note'])
    df.to_csv(f"{name_hotel[0]}_reviews.csv")
    return df
```

```python
get_data_hotel('https://www.tripadvisor.fr/Hotel_Review-g187147-d1631156-Reviews-Hotel_Joyce_Astotel-Paris_Ile_de_France.html')
```
